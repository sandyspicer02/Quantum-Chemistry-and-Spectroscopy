```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
init_printing()
%matplotlib inline
```

```python
n=Symbol('n',integer=True,positive=True)
L=Symbol("L",real=True,positive=True)
hbar=Symbol('hbar',real=True,positive=True)

f = simplify( (hbar*n*pi/L) * sqrt((L**2)/3 - (L**2)/(2*pi**2* n**2) - (L**2)/4) )
f
```

```python
p = plot(f/hbar,(n,0.6,2),show=False)

p.extend(plotting.plot_parametric(n,0.5,(n,0.6,2),show=False))
p[1].line_color='red'
p.show()
```

```python
# Ground State Wavefunction

f = (1/(0.1077*pi**.5))**(.5) * E**-(((x/0.1077)**2)/2)
N(f)
```

```python
plot(f,(x,-.5,.5))
```

```python
# Ground State Wavefunction Squared

plot(f**2,(x,-.5,.5))
N(f**2)
```
